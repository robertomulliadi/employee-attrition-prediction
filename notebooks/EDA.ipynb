# EDA

```r
library(dplyr, quietly = TRUE)
library(ggplot2, quietly = TRUE)
library(readr, quietly = TRUE)
library(tidyverse, quietly = TRUE)
library(janitor, quietly = TRUE)  
```

```r
data <- read_csv("../data/ibm_employee_data.csv")
data <- data |> clean_names()

# glimpse(data)
head(data, 5)
tail(data, 5)
```

```r
data |>
  count(attrition) |>
  mutate(prop = n / sum(n))

data |>
  ggplot(aes(x = attrition, fill = attrition)) +
  geom_bar() +
  labs(title = "Attrition Distribution")
```

### Visualizing Attrition Across Key Features

```r
# By Job Role

data |>
  ggplot(aes(x = job_role, fill = attrition)) +
  geom_bar(position = "fill") +
  coord_flip() +
  labs(title = "Attrition Rate by Job Role", y = "Proportion")
```

```r
# By Department

data |>
  ggplot(aes(x = department, fill = attrition)) +
  geom_bar(position = "fill") +
  labs(title = "Attrition by Department", y = "Proportion")
```

```r
# By Overtime

data |>
  ggplot(aes(x = over_time, fill = attrition)) +
  geom_bar(position = "fill") +
  labs(title = "Attrition vs Overtime")
```

```r
# By Age

data |>
  ggplot(aes(x = attrition, y = age, fill = attrition)) +
  geom_boxplot() +
  labs(title = "Attrition by Age")
```

```r
# By monthly income

data |>
  ggplot(aes(x = attrition, y = monthly_income, fill = attrition)) +
  geom_boxplot(alpha = 0.7) +
  labs(title = "Monthly Income by Attrition") +
  theme_minimal()
```

```r
# Tenure-based attrition

data |>
  ggplot(aes(x = years_at_company, fill = attrition)) +
  geom_histogram(binwidth = 1, position = "dodge") +
  labs(title = "Attrition by Years at Company", x = "Years at Company") +
  theme_minimal()
```

```r

numeric_df <- data |> select_if(is.numeric)
corr_matrix <- cor(numeric_df, use = "complete.obs")

# Convert matrix to long format for ggplot
corr_long <- as.data.frame(as.table(corr_matrix))
colnames(corr_long) <- c("Var1", "Var2", "Correlation")

ggplot(corr_long, aes(x = Var1, y = Var2, fill = Correlation)) +
  geom_tile(color = "white") +
  scale_fill_gradient2(low = "#D73027", high = "#1A9850", mid = "white",
                       midpoint = 0, limit = c(-1, 1), space = "Lab",
                       name = "Correlation") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, vjust = 1, hjust = 1)) +
  labs(title = "Correlation Heatmap of Numeric Features",
       x = "", y = "")

```

```r
# corr > 0.7
high_corr_pairs <- corr_long |>
  filter(Var1 != Var2) |>                                 # Exclude self-correlations
  filter(abs(Correlation) > 0.7) |>                      # Set your correlation threshold
  mutate(pair = map2_chr(Var1, Var2, ~ paste(sort(c(.x, .y)), collapse = "_"))) %>%
  distinct(pair, .keep_all = TRUE) |>
  arrange(desc(abs(Correlation))) |>
  select(Var1, Var2, Correlation)
print(high_corr_pairs)
```

```r
attrition_means <- data |>
  group_by(attrition) |>
  summarise(across(where(is.numeric), ~ mean(.x, na.rm = TRUE))) |>
  pivot_longer(-attrition, names_to = "feature", values_to = "mean_value")

attrition_means |>
  ggplot(aes(x = reorder(feature, -mean_value), y = mean_value, fill = attrition)) +
  geom_col(position = "dodge") +
  coord_flip() +
  labs(title = "Mean Values of Numeric Features by Attrition Status",
       x = "Feature",
       y = "Mean Value") +
  scale_fill_manual(values = c("#E69F00", "#56B4E9")) +
  theme_minimal()
```

## Summary of EDA Insights

## 🗂️ Dataset Overview

- **Total records:** 1,470
- **Total features:** 35
- **Target variable:** `Attrition` (Yes/No)
- Features include a mix of:
  - **Categorical:** job role, department, education, overtime, etc.
  - **Ordinal:** job level, performance rating
  - **Numeric:** age, monthly income, years at company, etc.

---

## ⚖️ Attrition Class Distribution

| Status | Count | Proportion |
|--------|-------|------------|
| No     | 1,233 | 83.9%      |
| Yes    | 237   | 16.1%      |

This imbalance indicates a **minority class problem**, requiring careful evaluation metrics during modeling (e.g., precision, recall, F1-score).

---

## 🔍 Key Categorical Variable Insights

### 🧪 Department
- **Highest attrition:** Sales and Human Resources
- **Lowest attrition:** Research & Development
- Employees in **customer-facing roles** may face higher pressure, explaining elevated churn.

### 👔 Job Role
- **Sales Representatives** and **Laboratory Technicians** had the **highest attrition**.
- **Research Directors** and **Manufacturing Directors** had the **lowest**.
- May relate to job satisfaction or stability in leadership roles.

### ⏱️ Overtime
- A **strong predictor of attrition**.
- Those who worked overtime had **much higher rates** of attrition.
- This relationship was **statistically significant** (Chi-square test p < 0.001).

### ❤️ Job Satisfaction & Work-Life Balance
- Lower satisfaction scores correspond to **higher attrition rates**.
- Employees with **poor work-life balance** had higher churn.
- Differences in attrition rates by satisfaction level were **statistically significant**.

---

## 📈 Numeric Feature Analysis

### 📌 Age
- Employees who left were generally **younger**.
- Mean age for leavers: ~33  
- Mean age for stayers: ~37  
- **t-test** confirms this difference is **statistically significant** (p < 0.001).

### 📌 Monthly Income
- Attrition negatively correlated with income.
- Leavers had **lower average income**.
- Boxplot showed a significant income gap; **t-test p < 0.001**.

### 📌 Years at Company
- Strong attrition peak at **0–3 years tenure**.
- Indicates many employees **leave early** in their tenure.
- Histogram visualized a right-skewed distribution among leavers.

### 📌 Total Working Years, Promotions, Manager Tenure
- Leavers:
  - Had **fewer total years of experience**
  - Were **less likely to be promoted**
  - Had **shorter relationships with managers**
- These trends were consistent and statistically significant across groups.

---

## 🔄 Correlation Insights

- `total_working_years` and `years_at_company`: **high positive correlation**
- `job_level` and `monthly_income`: **strong correlation**
- These may indicate **redundancy**, requiring feature selection or dimensionality reduction.
- Correlations visualized with a clean `ggplot2` heatmap (|r| > 0.7 pairs flagged).

---

## 📊 Grouped Feature Means by Attrition

Averages for numeric features were compared by attrition group:

| Feature             | Stayed (No) | Left (Yes) |
|---------------------|-------------|------------|
| Age                 | 37.1        | 33.6       |
| Monthly Income      | 6832        | 4787       |
| Years at Company    | 7.4         | 5.1        |
| Years Since Last Promotion | 2.2   | 1.5        |
| Years with Manager  | 4.2         | 2.8        |

These patterns support modeling with **logistic regression, tree-based models, or gradient boosting**, and justify feature importance exploration using SHAP.
